```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/mayank8200/Car-Price-Prediction/master/car_data.csv")
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,[1,3,4,6]].values
y=dataset.iloc[:,2].values
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
X[:,2]=lb.fit_transform(X[:,2])
lb1=LabelEncoder()
X[:,3]=lb1.fit_transform(X[:,3])
```

```python
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid={'n_estimators':[200,300]}
rf=RandomForestRegressor().fit(X,y)
gridlog=GridSearchCV(rf,param_grid,cv=3)


gridlog.fit(X_train,y_train)
print(gridlog.best_params_)
```

```python
accuracy=gridlog.score(X_test,y_test)
print(accuracy*100,'%')
```

```python
y_pred=gridlog.predict(X_test)
```
